## Noise Injection (Column-wise augmentation, Time domain)

image.png

```python
def white_noise_injection(series, pct=0.1): #percentage of noise
    """
    inputs
        series
            a pandas Series, must have DatetimeIndex
        pct
            a percentage of noise injection, within interval [0.0, 1.0] (default 0.1)
    output
        series
            a pandas Series with DatetimeIndex
    """
    temp = series.copy()

    def __filter_out_1sigma(arr):
        mu, sigma = np.mean(arr), np.std(arr)
        return arr[((mu - sigma) <= arr) & (arr <= (mu + sigma))]

    sr_len = len(temp)
    idx = np.arange(sr_len)
    _dif = np.diff(temp.values)
    dif = __filter_out_1sigma(_dif)

    _mean, _std = np.mean(dif), np.std(dif)
    pos = np.random.choice(idx, int(sr_len * pct))
    noise = np.random.normal(_mean, _std, size=len(temp))[pos]
    temp.iloc[pos] += noise
    return temp
```

```python
noise_ratio = np.arange(0.1,1,0.1)


# 4행 4열의 서브플롯 생성
fig, axs = plt.subplots(9, 10, figsize=(30, 30))

# 데이터를 서브플롯에 그리기
for i in range(9):
    for j in range(10):
        axs[i, j].plot(df[df.columns[j]],label='og')
        axs[i, j].plot(white_noise_injection(df[df.columns[j]],noise_ratio[i]),label='aug')
        axs[i, j].set_title(f'{df.columns[j]} Noise Ratio {noise_ratio[i]}')
        axs[i, j].legend()

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

## Fourier Transform (Column-wise augmentation, Frequency domain)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft

def fourier_time_series_augmentation(time_series, segment, magnitude_var=0.1, phase_var=0.1, mu_A=0, q_A=1):
    # Transform the time series into the frequency domain using FFT
    freq_domain = fft(np.array(time_series))

    # Magnitude Augmentation (Amplitude)
    real_part = np.real(freq_domain)
    imag_part = np.imag(freq_domain)

    #Amplitude
    magnitude = np.sqrt(real_part**2 + imag_part**2)

    # Apply perturbations in the magnitude spectrum
    magnitude_perturbation = np.random.normal(mu_A, magnitude_var, size=freq_domain.shape)
    magnitude[segment] += q_A * magnitude_perturbation[segment]

    # Phase Augmentation
    phase = np.angle(freq_domain)

    # Apply perturbations in the phase spectrum
    phase_perturbation = np.random.normal(0, phase_var, size=freq_domain.shape)
    phase[segment] += phase_perturbation[segment]

    # Reconstruct the frequency domain data
    augmented_freq_domain = magnitude * np.exp(1j * phase)

    # Transform the data back to the time domain using the inverse FFT
    augmented_time_series = ifft(augmented_freq_domain)

    return augmented_time_series.real
```

```python
def random_segment(length, segment_length):
    """
    Generate a random segment within the given length.
    Args:
    - length (int): Total length of the time series.
    - segment_length (int): Length of the segment to be generated.

    Returns:
    - segment (range): A range object representing the segment.
    """
    start = np.random.randint(0, length - segment_length)
    end = start + segment_length
    return range(start, end)

def fourier_multivariate_time_series_augmentation_random_segment(multivariate_time_series, segment_length, magnitude_var=0.1, phase_var=0.1, mu_A=0, q_A=1):
    # Number of variables in the multivariate time series
    num_variables = multivariate_time_series.shape[1] #10

    # Generate a random segment
    segment = random_segment(multivariate_time_series.shape[0], segment_length) #27,7

    # Initialize an array to hold the augmented time series
    augmented_multivariate_time_series = pd.DataFrame(columns=multivariate_time_series.columns,index=multivariate_time_series.index)

    # Apply Fourier augmentation for each variable
    for i in range(num_variables):
        augmented_multivariate_time_series.iloc[:, i] = fourier_time_series_augmentation(
            multivariate_time_series.iloc[:, i], segment, magnitude_var, phase_var, mu_A, q_A
        )

    return augmented_multivariate_time_series
```

```python
# Example usage with random segment
segment_length = 7 # Define the length of the segment for augmentation
augmented_multivariate_time_series_random_segment = fourier_multivariate_time_series_augmentation_random_segment(multivariate_time_series= df, segment_length = segment_length)

# Plot the augmented multivariate time series for the first variable with random segment
plt.figure(figsize=(15, 5))
plt.plot(augmented_multivariate_time_series_random_segment.iloc[:, 2])
plt.title("Augmented Time Series with Random Segment (Variable 1)")
plt.show()
```

```python
noise_aug = [df,aug_01,aug_02,aug_03,aug_04,aug_05,aug_06,aug_07,aug_08,aug_09]
final_aug = []

for elem in noise_aug:
    segment_length = 7 # Define the length of the segment for augmentation
    for i in range(4):
        fourier_aug = fourier_multivariate_time_series_augmentation_random_segment(elem,segment_length)
        final_aug.append(fourier_aug)
```
