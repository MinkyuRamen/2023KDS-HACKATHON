```python
date_range = pd.date_range(start='2023-01-01', end='2060-01-01', freq='YS')

# 데이터프레임 생성
born = pd.DataFrame({'Date': date_range})
born.set_index('Date', inplace=True)
born['출생아수'] = ''
```

```python
born['출생아수']['2023-01-01'] = 100000
for i in range(1, len(born)):
    born['출생아수'][i] = born['출생아수'][i-1]-1000*rd.random()
```

```python
born.to_csv('../raw_data/출생아수_example.csv')
```
