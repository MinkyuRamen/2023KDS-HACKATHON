```python
insix = cust[[cust_col[1], cust_col[3], cust_col[6], cust_col[7], cust_col[10], cust_col[11]]]
insix
```

```python
insix.columns = insix.columns.to_flat_index().map('_'.join)
```

```python
insix.replace('-', 0, inplace=True)
insix.index = pd.to_datetime(insix.index)
insix['year'] = insix.index.year.astype(str)
```

```python
insix.set_index('year', inplace=True)
insix = insix[~insix.index.duplicated(keep='first')]
```

```python
insix.index = pd.to_datetime(insix.index.astype(str))
insix
```
